```python
import pandas as pd
```

```python
text = open(r"C:\Users\Maria Gubina\Desktop\xyidxstats.txt").read().splitlines()
text = {text[i] : [int(text[i+1][2:].split()[0]), int(text[i+2][2:].split()[0])] for i in range(0, len(text), 3)}
text
```

```python
dct = {str(i): 'M' if i in (3, 5, 6, 12, 13, 15) else 'F' for i in range(1, 16)}
dct
```

```python
df = pd.DataFrame(text).transpose()
df.rename({0: 'X', 1: 'Y'}, inplace=True, axis="columns")
df['ratio_percent'] = round(df.Y / df.X * 100, 2)
df['sex'] = df.index.map(lambda x: dct[x.split('_')[1][1:]])
df.to_csv(r'C:\Users\Maria Gubina\Desktop\xyidxstats.csv')
df
```
